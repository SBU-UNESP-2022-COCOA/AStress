```python
#For calculating large-scale structure and lensing results yourself, get a power spectrum
#interpolation object. In this example we calculate the CMB lensing potential power
#spectrum using the Limber approximation, using PK=camb.get_matter_power_interpolator() function.
#calling PK(z, k) will then get power spectrum at any k and redshift z in range.

nz = 100 #number of steps to use for the radial/redshift integration
kmax=10  #kmax to use


## assume to be the same for all models 
# #For Limber result, want integration over \chi (comoving radial distance), from 0 to chi_*.
# #so get background results to find chistar, set up a range in chi, and calculate corresponding redshifts
LCDM= camb.get_background(pars0)
chistar = LCDM.conformal_time(0)- LCDM.tau_maxvis
chis = np.linspace(0,chistar,nz)
zs=LCDM.redshift_at_comoving_radial_distance(chis)
#Calculate array of delta_chi, and drop first and last points where things go singular
dchis = (chis[2:]-chis[:-2])/2
chis = chis[1:-1]
zs = zs[1:-1]

#Get the matter power spectrum interpolation object (based on RectBivariateSpline). 
#Here for lensing we want the power spectrum of the Weyl potential.
PK_0_Weyl = camb.get_matter_power_interpolator(pars0, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_Weyl,var2=model.Transfer_Weyl, zmax=zs[-1])

PK_1_Weyl = camb.get_matter_power_interpolator(pars1, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_Weyl,var2=model.Transfer_Weyl, zmax=zs[-1])

PK_2_Weyl = camb.get_matter_power_interpolator(pars2, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_Weyl,var2=model.Transfer_Weyl, zmax=zs[-1])

PK_3_Weyl = camb.get_matter_power_interpolator(pars3, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_Weyl,var2=model.Transfer_Weyl, zmax=zs[-1])

PK_4_Weyl = camb.get_matter_power_interpolator(pars4, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_Weyl,var2=model.Transfer_Weyl, zmax=zs[-1])

#Have a look at interpolated power spectrum results for a range of redshifts
#Expect linear potentials to decay a bit when Lambda becomes important, and change from non-linear growth
plt.figure(figsize=(8,5))
k=np.exp(np.log(10)*np.linspace(-4,2,200))
#zplot = [0, 0.5, 1, 4 ,20]
zplot = [0.5]
for z in zplot:
    plt.loglog(k, PK_0_Weyl.P(z,k),label='LCDM, g0=0',color="black")
    
for z in zplot:
    plt.loglog(k, PK_1_Weyl.P(z,k),label='g0=0.5',color="indianred")
    
for z in zplot:
    plt.loglog(k, PK_2_Weyl.P(z,k),label='g0=1.0',color="darkred")
    
for z in zplot:
    plt.loglog(k, PK_3_Weyl.P(z,k),label='g0=-0.5',color="lightskyblue")
    
for z in zplot:
    plt.loglog(k, PK_4_Weyl.P(z,k),label='g0=-1.0',color="steelblue")
    
    
    
    
plt.xlim([1e-4,1.5])
plt.xlabel('k Mpc')
plt.ylabel('$P_\Psi\, Mpc^{-3}$')
plt.legend(loc='lower right',fontsize='x-large')
plt.title('z=0.5')
```

```python
# #Test relation of P_phi and Pk matter, in the presence of AStress, check 1506.06313 for example
#testing: the factor doesn't seem to be correct


PK_0_matter = camb.get_matter_power_interpolator(pars0, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_tot,var2=model.Transfer_tot, zmax=zs[-1])

PK_4_matter = camb.get_matter_power_interpolator(pars4, nonlinear=True, 
    hubble_units=False, k_hunit=False, kmax=kmax,
    var1=model.Transfer_tot,var2=model.Transfer_tot, zmax=zs[-1])



factor_0 = PK_0_Weyl.P(0 , 0.001) / PK_0_matter.P(0 , 0.001)
factor_4 = PK_4_Weyl.P(0 , 0.001) / PK_4_matter.P(0 , 0.001)


plt.figure(figsize=(8,5))
k=np.exp(np.log(10)*np.linspace(-4,2,200))
#zplot = [0, 0.5, 1, 4 ,20]
zplot = [0.0]
for z in zplot:
    plt.loglog(k, PK_0_Weyl.P(z,k) ,label='LCDM, Weyl',color="black",linestyle='solid')
    plt.loglog(k, PK_0_matter.P(z,k) * factor_0 ,label='LCDM, matter',color="black",linestyle='dashdot')
    
for z in zplot:
    plt.loglog(k, PK_4_Weyl.P(z,k) ,label='g0=-1.0, Weyl',color="red",linestyle='solid')
    plt.loglog(k, PK_4_matter.P(z,k) * factor_4 ,label='g0=-1.0, matter',color="red",linestyle='dashdot')
    
    
    
    
plt.xlim([1e-4,1e0])
plt.xlabel('k Mpc')
plt.ylabel('$P_\Psi\, Mpc^{-3}$')
plt.legend(loc='lower right',fontsize='x-large')
plt.title('z=0.0')
```
